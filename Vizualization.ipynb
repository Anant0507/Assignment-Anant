```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
xpts = np.array([0,3,6,0,6,0])
ypts = np.array([0,6,0,4,4,0])
plt.plot(xpts,ypts,color="g",marker="*")
plt.show()
```

```python
xpts = np.array([1,3,2,7])
ypts = np.array([0,9,2,3])
plt.plot(xpts,ypts,"*")
plt.show()
```

```python
ypts = np.array([0,9,8,3])
plt.plot(xpts,ypts,color="b",marker="o")
plt.show()
```
